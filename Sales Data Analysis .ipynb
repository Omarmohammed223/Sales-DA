```python
# import python libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#load data 
df = pd.read_csv("Sales Data Analysis.csv", encoding='ISO-8859-1')
df.head()
```

```python
df.info()
```

```python
# drop blank columns
df.drop(['Status','unnamed1'] , axis = 1 , inplace=True)
```

```python
# check for null values
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
# change data type
df['Amount'] = df['Amount'].astype(int)
df['Amount'].dtypes
```

```python
# describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)
df.describe()
```

```python
# use describe() for specific columns
df[['Age','Orders','Amount']].describe()
```

## Exploratory Data Analysis
## Gender

```python
# plotting a bar chart for Gender and it's count
ax = sns.countplot(x='Gender', data=df)
sns.set(rc={'figure.figsize':(20,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount
sales_gender = df.groupby(['Gender'],as_index=False).sum().sort_values(by='Amount',ascending=False)
sns.set(rc= {'figure.figsize':(20,8)})
sns.barplot(x='Gender',y='Amount',data = sales_gender);
```

## From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

## Age

```python
ax = sns.countplot(x='Age Group',hue='Gender',data=df)
sns.set(rc={'figure.figsize':(20,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index=False).sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(x='Age Group',y='Amount',data=sales_age);
```

## From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

## State

```python
# total number of amount/sales from top 10 states
sales_state = df.groupby('State',as_index=False).sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(x='State',y='Orders',data=sales_state);
plt.title("Number of amount/sales by Top 10 States");
```

```python
# total number of amount/sales from top 10 states
sales_state = df.groupby('State',as_index=False).sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20, 8)})
sns.barplot(x='State',y='Amount',data=sales_state);
plt.title("Top 10 States with the Most Sales");
```

## From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively


## Marital Status

```python
ax = sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(20,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'],as_index=False).sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(x='Marital_Status',y='Amount',data=sales_state,hue='Gender');
```

## From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

## Occupation

```python
ax = sns.countplot(x='Occupation',data  = df)
sns.set(rc= {'figure.figsize':(20,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_ocp = df.groupby(['Occupation'],as_index=False).sum().sort_values(by='Amount',ascending=False)
sns.set(rc= {'figure.figsize':(20,8)})
sns.barplot(x='Occupation',y='Amount',data=sales_ocp);
```

## From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

## Product Category

```python
ax = sns.countplot(x='Product_Category',data=df)
sns.set(rc= {'figure.figsize': (30,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_pc = df.groupby(['Product_Category'],as_index=False).sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc= {'figure.figsize':(20,8)})
sns.barplot(x='Product_Category',y='Amount',data = sales_pc);
```

## From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

## Product_ID

```python
sales_pcid = df.groupby(['Product_ID'],as_index=False).sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc= {'figure.figsize':(20,8)})
sns.barplot(x='Product_ID',y='Orders',data=sales_pcid);
```

## Conclusion:
Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
